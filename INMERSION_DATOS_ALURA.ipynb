#**APLICANDO MODELO DE MACHINE LEARNING**
Modelo predictivo clásico, supervisado de Regresión Lineal
Vamos a estimar el precio de los inmuebles.

```python
#scikit learn biblioteca interesante de python para los modelos de machine learning. Nuestro paquete de regresion lineal.

from sklearn.linear_model import LinearRegression
#importamos otra librería para validar nuestro modelo, no solo nos interesa el modelado. Necesitamos separar el conjunto de datos para entrenarlo y datos de prueba
# para aplicar metricas y entender si nuestro modelo funciona bien.
from sklearn.model_selection import train_test_split
#vamos a poner los atributos como variables de entrada en el modelo y vamos a obtener solo una variable de respuesta que es el precio en millones de los inmuebles (lo que queremos predecir)
#Es bueno siempre generar una baseline para arrancar en el modelo, y ver si a medida que avanzamos con el modelo lo vamos mejorando.
#En esta baseline tomamos una matriz como entrada y un array como respuesta.

X = datos_ml[['COD_UPZ_GRUPO']] #una matriz con mi dataframe
y = datos_ml['Valor'] #un array con mi pandas series
#Ahora hacemos un split de ambos dataset que hemos creado. 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state =99)
```

```python
#lo que hicimos fue crear el modelo y destinando 25% para test y 75% para prueba
X_train
```

```python
# Tanto para x e y de test se usó el 25% que dan los 2079 registros y para entrenar las 6234 filas que son el 75% del dataset total= 8313 filas
```

```python
#instanciamos el modelo
modelo = LinearRegression()
```

```python
#Ajustando el modelo
modelo.fit(X_train, y_train)
```

```python
# Validando modelo con los datos de prueba
```

```python
y_predict_test =modelo.predict(X_test)
```

```python
#El MAE error absoluto medio, es una metrica que nos arroja un rango de error en el que se maneja nuestro modelos. Por ejemplo si
#nuestro inmueble cuesta 50 millones, podemos tener un margen de error de 20 millones mas o 20 millones de menos despues de que introduzcas tus valores en el modelo
#La idea es reducir estos valores para acercarnos lo maximo al valor real.
#R^2 R cuadrado es el coeficiente de determinación, varía entre 0 y 1 y nos indica que tan bien está trabajando nuestro modelo. Que se acerque cada vez más a 1 es lo óptimo.
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
baseline_mae = mean_absolute_error(y_test, y_predict_test)
baseline_r2 = r2_score(y_test,y_predict_test)
print(baseline_mae,baseline_r2)
#aquí instanciamos el error absoluto medio, comparando el "y" de prueba(que el modelo nunca vió) con el "y" predicho con nuestro machine learning  
```

```python
#Error absoluto medio: el precio del inmueble que se predice puede tener un error de +- 187.32 millones de COP.
```

#**MEJORAMOS NUESTRO MODELO AÑADIENDO MAS ATRIBUTOS**

```python
X = datos_ml[['COD_UPZ_GRUPO','Habitaciones','Banos','CONJUNTO_CERRADO','SALARIO_ANUAL_MI','TIENE_ESCRITURA']]
y = datos_ml['Valor']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state= 99)

modelo_1 = LinearRegression()
modelo_1.fit(X_train, y_train)
y_predict_test = modelo_1.predict(X_test)
y_predict_train = modelo_1.predict(X_train)

mae_test = mean_absolute_error(y_test,y_predict_test)
r2_test = r2_score(y_test, y_predict_test)

mae_train = mean_absolute_error(y_train, y_predict_train)
r2_train = r2_score(y_train, y_predict_train)
print(mae_test, r2_test)
print(mae_train, r2_train)
```
